```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

```python
X_train = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_train = np.sin(X_train)

X_train = X_train.reshape(-1, 1)
```

```python
model = keras.Sequential(
    [
        layers.Dense(32, activation="linear"),
        layers.Dense(16, activation="tanh"),
        layers.Dense(1),
    ]
)

model.compile(optimizer="adam", loss="mean_squared_error")

model.fit(X_train, y_train, epochs=1000, verbose=0)

X_test = np.linspace(-2 * np.pi, 2 * np.pi, 1000).reshape(-1, 1)
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(X_test, np.sin(X_test), label="True Sine Wave", color="blue")
plt.plot(
    X_test,
    y_pred,
    label="Neural Network Approximation",
    color="red",
    linestyle="dashed",
)
plt.title("Sine Wave Approximation using Neural Network")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.legend()
plt.grid()
plt.show()
```
