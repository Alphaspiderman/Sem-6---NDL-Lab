# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Linear

```python
def linear(x):
    return 4 * x


x = 1
linear(x)
```

### Test & Plot

```python
x = np.linspace(-10, 10, 50)
p = linear(x)

plt.xlabel("x")
plt.ylabel("Linear(x)")
plt.plot(x, p)
plt.show
```

### Sigmoid

```python
def sig(x):
    return 1 / (1 + np.exp(-x))


x = 1
print("Applying Sigmoid Activation Function on (%.1f) gives %.1f" % (x, sig(x)))
```

### Test & Plot

```python
x = np.linspace(-10, 10, 50)
p = sig(x)

plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x, p)
plt.show
```

### Tanh

```python
def tanh(x):
    return np.tanh(x)


x = 1
print("Applying Tanh Activation Function on (%.1f) gives %.1f" % (x, tanh(x)))
```

### Test & Plot

```python
x = np.linspace(-10, 10, 50)
p = tanh(x)

plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.plot(x, p)
plt.show
```

### ReLu

```python
def relu(x):
    return np.maximum(0, x)


x = 1
print("Applying ReLu Activation Function on (%.1f) gives %.1f" % (x, relu(x)))
```

### Test & Plot

```python
x = np.linspace(-10, 10, 50)
p = relu(x)

plt.xlabel("x")
plt.ylabel("ReLu(x)")
plt.plot(x, p)
plt.show
```

### Leaky ReLu

```python
def leaky_relu(x):
    return np.maximum(x, x * 0.01)


x = 1
print(
    "Applying Leaky ReLu Activation Function on (%.1f) gives %.1f" % (x, leaky_relu(x))
)
```

### Test & Plot

```python
x = np.linspace(-10, 10, 50)
p = leaky_relu(x)

plt.xlabel("x")
plt.ylabel("Leaky ReLu(x)")
plt.plot(x, p)
plt.show
```

### Softmax

```python
def softmax(x):
    exp_scores = np.exp(x)
    return exp_scores / np.sum(exp_scores)


x = 1
print("Applying Softmax Activation Function on (%.1f) gives %.1f" % (x, softmax(x)))
```

### Test & Plot

```python
x = np.linspace(-10, 10, 50)
p = softmax(x)

plt.xlabel("x")
plt.ylabel("SoftMax(x)")
plt.plot(x, p)
plt.show
```
